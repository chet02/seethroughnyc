```python
original_cols = PropertiesDF.columns

target_cols = ['BOROUGH', 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY',
       'TAX CLASS AT PRESENT', 'BLOCK', 'LOT', 'EASE-MENT',
       'BUILDING CLASS AT PRESENT', 'ADDRESS', 'APARTMENT NUMBER', 'ZIP CODE',
       'RESIDENTIAL UNITS', 'COMMERCIAL UNITS', 'TOTAL UNITS',
       'LAND SQUARE FEET', 'GROSS SQUARE FEET', 'YEAR BUILT',
       'TAX CLASS AT TIME OF SALE', 'BUILDING CLASS AT TIME OF SALE',
       'SALE PRICE', 'SALE DATE']

z_score_cols = []


for col in target_cols:
    
    new_col_name = col + "_zscore"
    

    PropertiesDF[new_col_name] = stats.stats.zscore(PropertiesDF[col] )
    

    PropertiesDF[new_col_name] = abs( PropertiesDF[new_col_name] )
    
 
    z_score_cols.append(new_col_name)


condition = PropertiesDF[z_score_cols] < 3
print(PropertiesDF.shape)


condition = condition.all(axis=1)

print('Before removal of outliers', PropertiesDF.shape)

PropertiesDF= PropertiesDF[condition]

print('After removal of outliers', PropertiesDF.shape)
```
